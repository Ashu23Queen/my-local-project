#### Discrete uniform distribution: Simulating a fair die roll

```python
# trying 100 die rolls and store in a list

die_rolls = [random.randint(1, 6) for x in range(1000000)] 
# randint(a, b) is a function from the random module that returns a random integer N such that a <= N <= b.

die_rolls
```

```python
# title pre-defined function to plot experimental uniform distributions

def plot_discrete(outcome_list):

    # plot
    plt.figure(figsize=(5,3))
    plt.hist(outcome_list, bins=range(1,8), align='left', rwidth=0.5, color='skyblue', edgecolor = 'black')
    plt.title("Discrete uniform distribution")
    plt.xlabel("outcome")
    plt.ylabel("Frequency")
    plt.grid(axis='y', linestyle = '--', alpha = 0.6)
    plt.xticks(range(1, 7))
    plt.show()
```

```python
import matplotlib.pyplot as plt # This is the standard import!
```

```python
# To call the function, pass the outcome list as an argument
plot_discrete(die_rolls) 
```

#### Continuous uniform distribution 

```python
# Simulate 100 values unoformly drawn from [0, 1]
uniform_continuous = [random.uniform(0, 1) for i in range(100000)]
```

```python
# title pre-defined function to plot experimental uniform distributions

def plot_continuous(outcome_list):

    # plot histogram
    plt.figure(figsize=(6, 4))
    plt.hist(outcome_list, bins=20, density=True, color='lightgreen', edgecolor = 'black')
    plt.title("Continuous Uniform Distribution: U(0, 1)")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.grid(axis='y', linestyle = '--', alpha = 0.6) 
    plt.show()
```

```python
plot_continuous(uniform_continuous)
```

#### Bernoulli & Binomial Distributions

```python
# Binomial Distribution
from scipy.stats import binom
```

```python
# Parameters
n = 3
p = 0.5  

# Define distribution
binom_dist = binom(n, p)

# Possible number of successes: 0 to n
x = list(range(n+1))

# PMF
pmf_vals = binom_dist.pmf(x)
```

```python
# title pre-defined function to plot binomial distribution

# Creating a function to plot binomial distribution from success probablity, list of possible successes, and pmf values

def plot_binom(n, pX, x_bin, pmf_bin):
     # plot histogram
    plt.figure(figsize=(8, 4))
    plt.bar(x, pmf_vals, color='skyblue', edgecolor = 'black')
    plt.title(f"Binomial Distribution: n={n}, p={p}")
    plt.xlabel("Number of Clicks {Successes}")
    plt.ylabel("Probability")
    plt.grid(axis='y', linestyle = '--', alpha = 0.6) 
    plt.show()

   
```

### Matplotlib to visualize a Binomial Distribution for "Product Defect Examples 

##### Visualizing Binomial Distribution with n=1 (Bernoulli Case)

Let's assume a defect rate (p) of 5%.

```python
# import matplotlib.pyplot as plt   already used
# from scipy.stats import binom     already used
import numpy as np
```

```python
# Parameters for our "single product" (Binomial with n=1)
n_single_product = 1 # Number of trials (single product)
p_defect = 0.05      # Probability of defect
```

```python
# Possible outcomes: 0 defects (not defective), 1 defect (defective)
k_values_single = np.arange(0, n_single_product + 1)

# Calculate the Probability Mass Function (PMF)
pmf_single = binom.pmf(k_values_single, n_single_product, p_defect)

print(f"Probabilities for a single product (n={n_single_product}, p={p_defect}):")
for k, prob in zip(k_values_single, pmf_single):
    print(f"  P(X={k} defects) = {prob:.4f}")
```

```python
# Plotting the PMF
plt.figure(figsize=(6, 4))
plt.bar(k_values_single, pmf_single, color=['skyblue', 'salmon'], edgecolor='black', alpha=0.7)

plt.title(f"Binomial Distribution (n={n_single_product}, p={p_defect}) - Single Product")
plt.xlabel("Number of Defects (k)")
plt.ylabel("Probability P(X=k)")
plt.xticks(k_values_single, ['Not Defective (0)', 'Defective (1)'])
plt.ylim(0, 1.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
